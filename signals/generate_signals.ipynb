```python
# imports
import numpy as np
from numpy import savetxt
import matplotlib.pyplot as plt
```

### X(t):

```python
X = np.loadtxt(open("X.csv", "rb"), delimiter=",")
X[1:]
```

```python
plt.plot(X[0],X[1])
plt.xlabel("Time (t)")
plt.ylabel("X(t)");
```

### Y(t) = Bsin(2* $pi$ *t):

```python
t = np.arange(0,2.1,0.1)
beta = np.random.uniform(low=0, high=1, size=(100,))
Y = np.zeros((101,21))
Y[0] = t
for i in range(1,101):
    Y[i] = beta[i-1] * np.sin(2*np.pi*t)
plt.plot(t,Y[1])
```

```python
savetxt("y.csv", Y, delimiter=",")
```

### Z(t) = X(t) x Y(t)

```python
Z = np.zeros((101,21))
Z[0] = t 
Z[1:] = np.multiply(X[1:,0:21],Y[1:])
plt.plot(t,Z[1])
plt.title("Z(t) Sample")
plt.xlabel("Time (t)")
plt.ylabel("Z(t)");
```

```python
savetxt("z.csv", Z, delimiter=",")
```
